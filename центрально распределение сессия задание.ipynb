```python
from scipy.stats import beta, norm
import matplotlib.pyplot as plt
import numpy as np
```

# создаем выборку по бета-распределению

```python
beta_rv=beta(5, 1)
sample=beta.rvs(5,1, size=1000)
```

# строим гистограмму распределения и теоретическую плотность

```python

x = np.linspace(0,1,100)
pdf= beta_rv.pdf(x)
plt.plot(x,pdf,lw=2, color='red', label='normal pdf')

plt.hist(sample,bins=10, normed=True, edgecolor = 'black')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.show()
```

# оценим выборочное среднее для разных n выборок

```python
# n==5
```

```python
n=5
sample=[0]*1000
for i in range(1000):
    viborka=beta.rvs(5,1, size=n) #cоздаем выборку размера n
    sentr=sum(viborka)/n #вычисляем среднее значение выборки
    sample[i]=sentr
    
D=5./252.  #дисперссия для бета распределения с коофицентами 5 и 1
D5=D/n  # дисперсия для нормального распределения
sigma=np.sqrt(D5) #cигма для нормальногор распределения
E=0.83 #мат ожидание бета распределения
norm_rv=norm(E,sigma) #задаем нормальное распределения для сигмы и мат ожидания бета распределения
x = np.linspace(0,1.5,100)
fun=norm_rv.pdf(x)
plt.plot(x, fun, label='norm') #строим график нормального распределения
plt.hist(sample,normed=True, edgecolor = 'black') #строем гистограмму средних значений
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.show()
```
